```python
# Reading the text file
def readTextFile(path):
    with open(path) as f:
        captions = f.read()
    return captions
```

```python
captions  = readTextFile("./Flickr8k_text/Flickr8k.token.txt")
captions = captions.split('\n')
print(captions[0])
print(captions[-2]) 
print(captions[-1]) #No caption in last line so we remove it
```

```python
captions=captions[:-1]
print(len(captions))
```

```python
first,second  = captions[0].split('\t')
print(first)
print(first.split(".")[0]) #image id
print(second) #its caption
```

```python
#making a dictionary that will map the image id with its corresponding captions
descriptions={}
for x in captions:
    first,second = x.split('\t')
    img_name = first.split(".")[0]
    if descriptions.get(img_name) is None:
        descriptions[img_name] = []
    descriptions[img_name].append(second)
```
